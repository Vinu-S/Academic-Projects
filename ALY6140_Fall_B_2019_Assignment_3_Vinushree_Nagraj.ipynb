```
#What is the average tip amount
#What is the average bill amount
#What is the average tip by gender
#What is the average tip on a sunday by gender
#Which party size gives the maximum tip as a percent of total bill amount
#If you were a restaurant manager, how might you use this data?
```

```
#average tip amount
average_tip=tips['tip'].mean()
print('Average tip amount is ' + str(average_tip))
```

```
#the average bill amount
average_bill=tips['total_bill'].mean()
print('Average bill amount is ' + str(average_bill))
```

```
#the average tip by gender
df_tip = tips.groupby(by=['sex'])['tip'].mean()
print(df_tip)
```

```
#the average tip on a sunday by gender
df = tips[tips['day']=='Sun']
average_df = df.groupby(by=(['sex']))['tip'].mean()
print(average_df)
```

```
#Which party size gives the maximum tip as a percent of total bill amount
tip_size = tips.groupby(['size']).max()
tip_size['percent'] = tip_size['tip']/tip_size['total_bill']*100
df_tip_size = tip_size.reset_index()
print(df_tip_size[df_tip_size.percent == df_tip_size.percent.max()])
```

#Using Matplotlib

```
#Generate a box plot that shows average bill amount by day of week(Matplotlib)
df = sns.load_dataset('tips')
df_1= df.boxplot(column='total_bill', by='day',patch_artist=True,grid=False)
df_1.set_xlabel("Day of the Week",
                fontsize=14)
df_1.set_ylabel("Total Bill",
                fontsize=14)
df_1.tick_params(labelsize=10)
plt.title('Bill amount by Day')
plt.suptitle('')
plt.show()
```

```
#Generate a box plot that shows average bill amount by gender(Matplotlib)

df_1= df.boxplot(column='total_bill', by='sex',patch_artist=True,grid=False)
df_1.set_xlabel("Gender",
                fontsize=14)
df_1.set_ylabel("Total Bill",
                fontsize=14)
df_1.tick_params(labelsize=10)
plt.title('Bill amount by Gender')
plt.suptitle('')
plt.show()
```

```
#Show the above two plots side by side(Matplotlib)
df=pd.DataFrame(tips)
fig, ax =plt.subplots(1,2, sharey=True)
df_1 = df.boxplot(by=("day"), column=["total_bill"],ax=ax[1],patch_artist=True,grid=False)
df_2 = df.boxplot(by=("sex"), column=["total_bill"],ax=ax[0],patch_artist=True,grid=False)
plt.suptitle('');
fig.show()
```

```
#linear relationship between tips and total_bill
sns.lmplot(x="total_bill", y="tip", data=tips);
```

```
#Do smoker/non-smoker give a higher tip as a percentage of total_bill (which line has a steeper slope)
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips ,palette="Set1");
```

```
#Although a smoker is more generous to pay more tips, the dots here are vaguely spread out.
#We only is that a non-smoker has a steeper slope and non-smokers are more consistent in paying tips and also is quite high.
```

```
#Are tips larger during lunch or dinner (as a percentage of total_bill)
sns.lmplot(x="total_bill", y="tip",  col="time", data=tips,aspect=.9);
```
